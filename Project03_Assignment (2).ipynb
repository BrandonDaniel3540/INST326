```python
# Solution - enter your code solution below
import calendar
from datetime import datetime
import random

class Caregiver:
    def __init__(self, name, phone, email, pay_rate, hours):
        self.name = name
        self.phone = phone
        self.email = email
        self.pay_rate = pay_rate
        self.hours = hours

    def get_name(self):
        return self.name

    def set_name(self, name):
        self.name = name

    def get_phone(self):
        return self.phone

    def set_phone(self, phone):
        self.phone = phone

    def get_email(self):
        return self.email

    def set_email(self, email):
        self.email = email

    def get_pay_rate(self):
        return self.pay_rate

    def set_pay_rate(self, pay_rate):
        self.pay_rate = pay_rate

    def get_hours(self):
        return self.hours

    def set_hours(self, hours):
        self.hours = hours

    def add_hours(self, additional_hours):
        self.hours += additional_hours


class AvailabilitySchedule:
    def __init__(self, caregiver):
        self.caregiver = caregiver

    def create_availability_schedule(self, availability):
        days = ["Sunday", "Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday"]
        for day in days:
            response = input(f"Is {self.caregiver.name} available on {day}? Enter 'Available' if yes: ")
            if response == "Available":
                availability[day]["Available"].append(self.caregiver.name)
            else:
                availability[day]["Unavailable"].append(self.caregiver.name)

        print(f"Availability updated for {self.caregiver.name}")


class CustomHTMLCalendar(calendar.HTMLCalendar):
    def __init__(self, events=None):
        super().__init__(calendar.SUNDAY)  # Start the week on Sunday
        self.events = events or {}  # Dictionary to hold events for specific dates

    def set_events(self, events):
        """Set or update events. Events should be a dictionary in the form {date: text}."""
        self.events = events

    def formatday(self, day, weekday):
        """Customize the format of each day."""
        if day == 0:
            # Day outside of the current month
            return '<td class="noday">&nbsp;</td>'
        else:
            # Add event text below the day number
            event_text = self.events.get(day, "")  # Get event for the day, if available
            return f'''
                <td class="{self.cssclasses[weekday]}">
                    <div>{day}</div>
                    <div style="font-size: small; color: gray;">{event_text}</div>
                </td>
            '''

    def formatmonth(self, year, month):
        """Generate the complete month's calendar as an HTML table."""
        return super().formatmonth(year, month)


# Helper function to assign shifts
def assign_shifts(day, availability):
    if availability[day]["Available"]:
        person = random.choice(availability[day]["Available"])
        for caregiver in caregivers:
            if caregiver.get_name() == person:
                caregiver.add_hours(6)
        return person
    else:
        return "No caregiver available"


# Main script
if __name__ == "__main__":
    # Define caregivers
    caregivers = [
        Caregiver("Mary", "577-414-4477", "maryb@gmail.com", 20, 0),
        Caregiver("Pamela", "339-546-0467", "pam1952@gmail.com", 0, 0),
        Caregiver("Paulene", "766-679-3324", "paulene1971@gmail.com", 0, 0),
        Caregiver("Michaela", "954-642-1751", "michaela1994@gmail.com", 0, 0),
        Caregiver("May", "606-458-4587", "mayc2002@gmail.com", 0, 0),
        Caregiver("Jessa", "464-237-7612", "jessmadalyn@gmail.com", 20, 0),
        Caregiver("Fae", "944-258-7552", "faed2000@gmail.com", 20, 0),
        Caregiver("Iola", "433-974-2129", "iola12423@gmail.com", 20, 0)
    ]

    # Define availability structure
    availability = {
        "Sunday": {"Available": [], "Unavailable": []},
        "Monday": {"Available": [], "Unavailable": []},
        "Tuesday": {"Available": [], "Unavailable": []},
        "Wednesday": {"Available": [], "Unavailable": []},
        "Thursday": {"Available": [], "Unavailable": []},
        "Friday": {"Available": [], "Unavailable": []},
        "Saturday": {"Available": [], "Unavailable": []},
    }

    # Collect availability for all caregivers
    for caregiver in caregivers:
        caregiver_schedule = AvailabilitySchedule(caregiver)
        caregiver_schedule.create_availability_schedule(availability)

    # Generate events for the calendar
    events = {}
    for day_number in range(1, 31):
        weekday = list(calendar.day_name)[(day_number - 1) % 7]  # Get the weekday name
        morning_shift = assign_shifts(weekday, availability)
        evening_shift = assign_shifts(weekday, availability)
        events[day_number] = f"7 AM - 1 PM: {morning_shift} <br> 1 PM - 7 PM: {evening_shift}"

    # Create a custom calendar instance
    custom_calendar = CustomHTMLCalendar(events)

    # Generate the calendar for November 2024
    calendar_html = custom_calendar.formatmonth(2024, 11)

    # Save to file
    with open("calendar.html", "w") as file:
        file.write(f"""
        <!DOCTYPE html>
        <html lang="en">
        <head>
            <meta charset="UTF-8">
            <meta name="viewport" content="width=device-width, initial-scale=1.0">
            <title>Custom Calendar</title>
            <style>
                table {{
                    width: 100%;
                    border-collapse: collapse;
                    margin: 20px 0;
                }}
                th, td {{
                    border: 1px solid #ddd;
                    text-align: center;
                    padding: 8px;
                }}
                th {{
                    background-color: #f4f4f4;
                }}
                td {{
                    height: 80px;
                    vertical-align: top;
                }}
                td div {{
                    margin: 2px 0;
                }}
            </style>
        </head>
        <body>
            {calendar_html}
        </body>
        </html>
        """)

    print("Custom calendar saved to 'calendar.html'")
    print("Custom pay report saved to 'pay_report.txt'")
    caregiver_calendar = ""
    for caregiver in caregivers:
        if caregiver.get_pay_rate() != 0:
           caregiver_calendar += f"Name: {caregiver.get_name()} | Hours: {caregiver.get_hours()} | Check for November 2024: ${caregiver.get_pay_rate() * caregiver.get_hours()}\n"
        else:
           caregiver_calendar += f"Name: {caregiver.get_name()} | Hours: {caregiver.get_hours()}\n"
    with open("pay_report.txt", "w") as file:
        file.write(caregiver_calendar)
```
